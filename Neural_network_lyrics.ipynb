# Neural Network for lyrics data

In this part I will deal with the construction of a neural network that determines the music genre based on the Spotify function and analyzes the sentiment of the lyrics.

## Preprocesing data 

### Import data 

```python
df_main = pd.read_json('spotify_lyrics_and_sentiment.json')
df_main.head()
```

```python
df = df_main[df_main.genre != ""]
df.head()
```

### Fit data 

```python
from sklearn.model_selection import train_test_split
import keras
from keras.models import Sequential, Model
from keras.layers import Conv2D, Flatten, Dense, Dropout, MaxPooling2D, Lambda
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.1, random_state = 0)
```

```python
# create model
model = Sequential([
Dense(2048,activation='sigmoid'),
Dense(1024,activation='sigmoid'),
Dense(512,activation='sigmoid'),
Dense(256,activation='relu'),
Dense(128,activation='relu'),
Dense(64,activation='relu'),
Dense(32,activation='relu'),
Dense(20)
])
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
batch_size = 20
epochs = 50

model.fit(X_train, y_train,
         batch_size = batch_size,
         epochs = epochs,
         validation_data = (X_test, y_test))
```

```python
scores = model.evaluate(X_test, y_test)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
model.save('mymodel_lyrics.h5')
```

```python
def classification_of_genre(y_pred):
    '''
    The function returns two values, the first is the music projection genre and
    the second is a vector containing the three best-matched musical genres.
    
    :param y_pred: matrix solution of model
    :return: best_genre(str), best_matching_genres(dict)
    '''
    
    kind_of_genre = ['Alternative', 'Blues', 'Christian', 'Comedy', 'Country', 'Dance',
       'Electronic', 'Folk', 'Hip Hop/Rap', 'Jazz', 'Latin', 'Musical',
       'New Age', 'Oldies', 'Pop', 'R&B', 'Rock', 'Ska', 'Soul', 'World']
    best_genre = []
    best_percent = []
    best_matching_genres = []
    for vec in y_pred:
        zip_genre = zip(kind_of_genre, vec)
        dictOfGenres = dict(zip_genre)
        sort_dictOfGenres = sorted(dictOfGenres.items(), key=lambda vec: vec[1], reverse=True)
        if(np.abs(sort_dictOfGenres[0][1] - sort_dictOfGenres[1][1]) < 0.15):
            if(np.abs(sort_dictOfGenres[1][1] - sort_dictOfGenres[2][1]) < 0.01):
                best_genre.append(sort_dictOfGenres[2][0])
                best_percent.append(sort_dictOfGenres[2][1])
            else:
                best_genre.append(sort_dictOfGenres[1][0])
                best_percent.append(sort_dictOfGenres[1][1])
        else:
            best_genre.append(sort_dictOfGenres[0][0])
            best_percent.append(sort_dictOfGenres[0][1])
        best_matching_genres.append(sort_dictOfGenres[:3])
    return best_genre, best_percent, best_matching_genres
```
